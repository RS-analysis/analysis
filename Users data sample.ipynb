```python
from pymongo import MongoClient
from bson.objectid import ObjectId
import matplotlib, math
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from itertools import groupby
from operator import itemgetter
from tabulate import tabulate
import functools
```

### Connecting to local mongo database

```python
host = 'localhost'
port = 27017
username = ''
password = ''
database = 'tcc-recsys-mongo'

client = MongoClient(host, port)
db = client[database]
```

```python
user_number = 1
users_data = []
for user in db.users.find():
    username = "user_{}".format(user_number)
    if user['facebook']['posts']:
        for facebook_post in user['facebook']['posts']:
            if 'message' in facebook_post.keys():
                entry = (username, facebook_post['message'], 'facebook')
                users_data.append(entry)
    if user['twitter']:
        for twitter_post in user['twitter']:
            entry = (username, twitter_post['text'], 'twitter')
            users_data.append(entry)
    user_number = user_number + 1
```

```python
users_data = pd.DataFrame(users_data)
users_data.columns = ['user', 'message', 'source']
users_data
```
